# 创建ndarray

```python
#创建数组最简单的办法就是使用array 函数
data1 = [6, 7.5, 8, 0, 1]
print('data1 type:',type(data1)) #list 没有dtype
arr1=np.array(data1)
print(arr1)
print('arr1 type:',arr1.dtype)
```

```python
#嵌套序列（比如由一组等长列表组成的列表）将会被转换为一个多维数组：

data2 = [[1, 2, 3, 4], [5, 6, 7, 8]]
arr2 = np.array(data2)
print(arr2)
print(arr2.shape)
```

```python
#除非特别说明,np.array会尝试为新建的这个数组推断出一个较为合适的数据类型。
#数据类型保存在一个特殊的dtype对象中。
print(arr1.dtype)
print(arr2.dtype)
```

```python
print(np.ones(shape=(2,3))) #默认数据类型为np.float64(浮点数)除非自己定义
print('*'*30)
print(np.zeros(10))
print('*'*30)
print(np.zeros((3, 6)))
print('*'*30)
print(np.empty((2, 3, 2)))
#np.empty返回的都是一些未初始化的垃圾值。
```

# ndarray的数据类型

# 花式索引

```python
#花式索引指的是利用整数数组进行索引。假设我们有一个8×4数组：
arr = np.empty((8, 4))
for i in range(8):
    arr[i] = i
print (arr)
```

# 数组转置和轴兑换

```python
#数组不仅有transpose方法，还有一个特殊的T属性
arr = np.arange(15).reshape((3, 5))
print(arr)
print("*"*20)
print(arr.transpose())
print("*"*20)
print(arr.T)
```

```python
#利用np.dot计算矩阵内积：
arr = np.random.randn(6, 3)
print (arr)
print("*"*40)
print (np.dot(arr.T, arr))
```

```python
#高维数组，transpose需要得到一个由轴编号组成的元组才能对这些轴进行转置
arr = np.arange(16).reshape((2, 2, 4))
print(arr)
print('==========分割线==============')
print(arr.transpose((1, 0, 2)))
#一个轴被换成了第二个，第二个轴被换成了第一个，最后一个轴不变
```

```python
#一个轴被换成了第二个，第二个轴被换成了第一个，最后一个轴不变
print(arr)
print('==========分割线==============')
print(arr.swapaxes(1, 2))
```

# 将条件逻辑表述为数组运算

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])

z=[(x if c else y) for x,y,c in zip(xarr,yarr,cond)] #等价于np.where(cond,x,y)
z
```

```python
np.where(cond,xarr,yarr)
```

```python
arr = np.random.randn(4, 4)
print(arr)
print(arr > 0)
print(np.where(arr > 0, arr, 5))  #np.where 可以将标量和数组结合起来
```

# 排序

#sort()函数返回输入数组的排序副本。
#np.sort(a, axis, kind, order)



```python
import numpy as np  
 
a = np.array([[3,7],[9,1]])  
print ('我们的数组是：')
print (a) 
print ('\n')  
print ('调用 sort() 函数：')  
print (np.sort(a))  
print ('\n')  
print ('沿轴 0 排序：')  
print (np.sort(a, axis =  0))  
print ('\n')  
# 在 sort 函数中排序字段 
dt = np.dtype([('name',  'S10'),('age',  int)]) 
a = np.array([("raju",21),("anil",25),("ravi",  17),  ("amar",27)], dtype = dt)  
print ('我们的数组是：')  
print (a) 
print ('\n')  
print ('按 name 排序：')  
print (np.sort(a, order =  'name'))
```

```python
#numpy.argsort()函数对输入数组沿给定轴执行间接排序，使用指定排序类型返回数据的索引数组。 
#这个索引数组用于构造排序后的数组。
arr = np.random.randn(5, 3)
print(arr)
print("*"*40) 
arr.sort(1) #对每一行进行排序，不考虑原始位置，产生一个新的副本
print(arr)
print("*"*40) 
arr.sort(0) #对每一列进行排序，不考虑原始位置，产生一个新的副本
print(arr)
```

```python
import numpy as np 
 
x = np.array([3,  1,  2])  
print ('我们的数组是：')  
print (x) 
print ('\n')  
print ('对 x 调用 argsort() 函数：' )
y = np.argsort(x)  #最小值的索引位置
print (y) 
print ('\n')  
print ('以排序后的顺序重构原数组：')  
print (x[y])  
print ('\n')  
print ('使用循环重构原数组：')  
for i in y:  
    print (x[i])
```

# 1.5 线性代数

```python
from numpy.linalg import inv, qr
 
X = np.random.randn(5, 5)
mat = X.T.dot(X)
print(inv(mat))
print(mat.dot(inv(mat)))
 
q, r = qr(mat)
print(r)
```
